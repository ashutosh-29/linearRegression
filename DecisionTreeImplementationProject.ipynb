```python
df = pd.DataFrame(iris.data)
df.columns = ['sl', 'sw', 'pl', 'pw']
```

```python
# we divide the different values of features in this format
# for less than (meanvalue + min)/2 is taken as 0
# for less than meanvalue is taken as 1
# for less than (meanvalue + max)/2 is taken as 2
# else it is taken as 3
def categorise(val, *compare) :
    if val < compare[0] :
        return 0
    elif val < compare[1] :
        return 1
    elif val < compare[2] :
        return 2
    else :
        return 3
```

```python
def new_feature(df, prevFeature) :
    minimum = df[prevFeature].min()
    maximum = df[prevFeature].max()
    mean = df[prevFeature].mean()
    compare = []
    compare.append((minimum + mean) / 2)
    compare.append(mean)
    compare.append((mean + maximum) / 2)
    return df[prevFeature].apply(categorise, args = (compare))
```

```python
df['sl_new'] = new_feature(df, 'sl')
df['sw_new'] = new_feature(df, 'sw')
df['pl_new'] = new_feature(df, 'pl')
df['pw_new'] = new_feature(df, 'pw')
```

```python
import math
```

```python
class TreeNode:
    def __init__(self, data, output) :
        # data represents the feature upon which the node was split when fitting the training data
        # data = None for leaf node
        self.data = data
        # children of a node are stored as a dicticionary with key being the value of feature upon which the node was split
        # and the corresponding value stores the child TreeNode
        self.children = {}
        # output represents the class with current majority at this instance of the decision tree
        self.output = output
        # index will be used to assign a unique index to each node
        self.index = -1
        
    def add_children(self, feature_value, value) :
        self.children[feature_value] = value
```

```python
class DecisionTreeClassifier:
    def __init__(self):
        # root represents the root node of the decision tree
        self.__root = None

    def __class_frequency(self, Y) : 
        # returns a dictionary with keys as unique values of Y(i.e no of classes) and the corresponding value as its frequency
        f = {}
        for i in Y:
            f[i] = f.get(i, 0) + 1
        return f

    def __entropy(self, Y) :
        # returns the entropy 
        freq_dict = self.__class_frequency(Y)
        entropy = 0
        for i in freq_dict:
            prob_xi = freq_dict[i] / len(Y)
            entropy -= prob_xi * math.log2(prob_xi)
            # i have used 2 as the log base
        return entropy

    def __gain_ratio(self, X, Y, feature):
        # returns the gain ratio
        # info_org represents entropy before splitting
        info_org = self.__entropy(Y)
        # info_f represents entropy after splitting upon the selected feature
        info_f = 0  
        split_info = 0
        values = set(X[:, feature])
        df = pd.DataFrame(X)
        # Adding Y values as the last column in the dataframe 
        df[df.shape[1]] = Y
        initial_size = df.shape[0] 
        for i in values:
            di = df[df[feature] == i]
            current_size = di.shape[0]
            info_f += (current_size / initial_size) * self.__entropy(di[di.shape[1] - 1])
            split_info -= (current_size / initial_size) * math.log2(current_size / initial_size)

        # to handle the case when split info = 0 which leads to division by 0 error
        if split_info == 0 :
            return math.inf

        info_gain = info_org - info_f
        gain_ratio = info_gain / split_info
        return gain_ratio


    def __decision_tree(self, X, Y, features, level, classes, feature_name):
        # returns the root of the Decision Tree(which consists of TreeNodes) built after fitting the training data
        # Here Nodes are printed as in PREORDER traversl
        # classes represents the different classes present in the set(Y) i.e., the classification
        # level is the depth of the tree
        # We split a node on a particular feature only once
        
        # base case
        # If the node consists of only 1 class i.e., pure class node
        if len(set(Y)) == 1:
            print("Level", level)
            output = None
            for i in classes :
                if i in Y :
                    output = i
                    print("Count of", i, "=", len(Y))
                else :
                    print("Count of", i, "=", 0)
                    
            print("Current Entropy is =  0.0")
            print("Reached leaf Node")
            print()
            return TreeNode(None, output)

        # If we have run out of features to split upon
        # In this case we will output the class with maximum frequency(i.e., we go with the majority)
        if len(features) == 0:
            print("Level",level)
            freq_dict = self.__class_frequency(Y)
            output = None
            max_count = -math.inf
            for i in classes:
                if i not in freq_dict:
                    print("Count of", i, "=", 0)
                else :
                    if freq_dict[i] > max_count :
                        output = i
                        max_count = freq_dict[i]
                    print("Count of", i, "=", freq_dict[i])
                    
            print("Current Entropy  is =", self.__entropy(Y))           
            print("Reached leaf Node")
            print()
            return TreeNode(None, output)

        
        # Finding the best feature to split upon
        max_gain = -math.inf
        final_feature = None
        for f in features :
            current_gain = self.__gain_ratio(X,Y,f)
            if current_gain > max_gain:
                max_gain = current_gain
                final_feature = f

        print("Level", level)
        freq_dict = self.__class_frequency(Y)
        output = None
        max_count = -math.inf

        for i in classes:
            if i not in freq_dict:
                print("Count of", i, "=", 0)
            else :
                if freq_dict[i] > max_count :
                    output = i
                    max_count = freq_dict[i]
                print("Count of", i, "=", freq_dict[i])

        
        print("Current Entropy is =", self.__entropy(Y))
        print("Splitting on feature", feature_name[final_feature], "with gain ratio ", max_gain)
        print()


            
        # unique_values represents the unique values of the feature selected
        unique_values = set(X[:, final_feature]) 
        df = pd.DataFrame(X)
        # Adding Y values as the last column in the dataframe
        df[df.shape[1]] = Y

        current_node = TreeNode(final_feature, output)

        # Now removing the selected feature from the list as we do not want to split on one feature more than once(in a given root to leaf node path)
        index  = features.index(final_feature)
        features.remove(final_feature)
        for i in unique_values:
            # Creating a new dataframe with value of selected feature = i
            di = df[df[final_feature] == i]
            # Segregating the X and Y values and recursively calling on the splits
            node = self.__decision_tree(di.iloc[:,0:di.shape[1]-1].values, di.iloc[:,di.shape[1]-1].values, features, level+1, classes, feature_name)
            current_node.add_children(i, node)

        # Add the removed feature     
        features.insert(index,final_feature)
        return current_node
    
    def fit(self, X, Y, feature_name) :
        # Fits the training data
        features = [i for i in range(len(X[0]))]
        classes = set(Y)
        level = 0
        self.__root = self.__decision_tree(X, Y, features, level, classes, feature_name)
        
    def __predict_for(self, data, node) :
        # predicts the class for a given testing point and returns the answer
        
        # We have reached a leaf node
        if len(node.children) == 0 :
            return node.output

        # represents the value of feature on which the split was made       
        val = data[node.data] 
        if val not in node.children :
            return node.output
        
        # Recursively call on the splits
        return self.__predict_for(data,node.children[val])

    def predict(self, X) :
        # This function returns Y predicted
        # X should be a 2-D np array
        Y = np.array([0 for i in range(len(X))])
        for i in range(len(X)):
            Y[i] = self.__predict_for(X[i], self.__root)
        return Y
    
    def score(self, X, Y):
        # returns the accuracy
        Y_pred = self.predict(X)
        count = 0
        for i in range(len(Y_pred)):
            if Y_pred[i] == Y[i]:
                count += 1
        scr = count / len(Y_pred)
        return scr
    
    
```

```python
from sklearn.model_selection import train_test_split
x = df.values
y = iris.target
x_train, x_test, y_train, y_test = train_test_split(x, y)
feature_name = list(df.columns)
clf = DecisionTreeClassifier()
clf.fit(x_train, y_train, feature_name)
print()
print("Score =", clf.score(x_test, y_test))
```
